```python
import numpy as np
import pandas as pd

EXPERIMENT = 'fashion_mnist'
DIM = [2,3,4]
STD = [0.01, 0.1, 0.2, 1.0]

D_Acc = np.zeros((len(DIM), len(STD)))
True_Std = np.zeros((len(DIM), len(STD)))
for d in range(len(DIM)):
    for s in range(len(STD)):
        load_file = 'results/discriminator/' + EXPERIMENT + '_dim_' + str(DIM[d]) + '_std_' + str(STD[s]) +'_.pickle'
        with open(load_file, 'rb') as file:
            load_data = pickle.load(file)
        D_Acc[d,s] = load_data['ortho'][0]
        True_Std[d,s] = load_data['ortho'][2]
```

```python
df_fashion = pd.DataFrame(data=D_Acc, index=DIM, columns=STD)
df_fashion
```

```python
EXPERIMENT = 'compass'
METHODS = ['rawper', 'per', 'plane', 'ortho']
DIM = [2]
STD = [0.01, 0.1, 1.0]

for METHOD in METHODS:
    D_Acc = np.zeros((len(DIM), len(STD)))
    True_Std = np.zeros((len(DIM), len(STD)))
    for d in range(len(DIM)):
        for s in range(len(STD)):
            load_file = 'results/discriminator/' + EXPERIMENT + '_dim_' + str(DIM[d]) + '_std_' + str(STD[s]) +'_.pickle'
            with open(load_file, 'rb') as file:
                load_data = pickle.load(file)
            D_Acc[d,s] = load_data[METHOD][0]
            True_Std[d,s] = load_data[METHOD][2]

    df_o = pd.DataFrame(data=D_Acc, index=DIM, columns=STD)
    df_std = pd.DataFrame(data=True_Std, index=DIM, columns=STD)

    from IPython.display import display, HTML

    print(EXPERIMENT + " " + METHOD)
    display(df_o)
#     display(df_std)
```
