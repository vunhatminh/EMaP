```python
PLOTCOL = ['H0_ortho', 'H1_ortho', 'H0_plane', 'H1_plane']
PLOT_RENAME = {PLOTCOL[0]: '$H_0$ Orthogonal',
               PLOTCOL[1]: '$H_1$ Orthogonal',
              PLOTCOL[2]: '$H_0$ Projection',
              PLOTCOL[3]: '$H_1$ Projection'}

PATH = 'result/bottleneck/'
SAVEPATH = 'result/images/'

df_dict = {}

for EXPERIMENT in ['mnist']:
    list_of_bigdfs = []
    for RADIUS in [0.001, 0.01, 0.1, 1.0]:
        load_files = PATH + EXPERIMENT + '_label_*_dim_' + str(2) + '_std_' + str(RADIUS) + '_.pickle'
        read_list = glob.glob(load_files)
        list_of_dfs = []
        for file in read_list:
            with open(file, 'rb') as file:
                d = pickle.load(file)
                for column in d:
                    new = d[[column]].copy()
                    if column in PLOTCOL:
                        new['measure'] = PLOT_RENAME[column]
                    else:
                        new['measure'] = column
                    new.columns = new.columns.str.replace(column, 'Bottleneck distance')
                    list_of_dfs.append(new)

        big_df = pd.concat(list_of_dfs)
        big_df['Noise level'] = RADIUS
        list_of_bigdfs.append(big_df)
    bigbig_df = pd.concat(list_of_bigdfs)
    df_dict[EXPERIMENT + str(2)] = bigbig_df

for EXPERIMENT in ['mnist']:    
    list_of_bigdfs = []
    for RADIUS in [0.001, 0.01, 0.1, 1.0]:
        load_files = PATH + EXPERIMENT + '_label_*_dim_' + str(3) + '_std_' + str(RADIUS) + '_.pickle'
        read_list = glob.glob(load_files)
        list_of_dfs = []
        for file in read_list:
            with open(file, 'rb') as file:
                d = pickle.load(file)
                for column in d:
                    new = d[[column]].copy()
                    if column in PLOTCOL:
                        new['measure'] = PLOT_RENAME[column]
                    else:
                        new['measure'] = column
                    new.columns = new.columns.str.replace(column, 'Bottleneck distance')
                    list_of_dfs.append(new)

        big_df = pd.concat(list_of_dfs)
        big_df['Noise level'] = RADIUS
        list_of_bigdfs.append(big_df)
    bigbig_df = pd.concat(list_of_bigdfs)
    df_dict[EXPERIMENT + str(3)] = bigbig_df
    
for EXPERIMENT in ['fashion_mnist']:
    list_of_bigdfs = []
    for RADIUS in [0.001, 0.01, 0.1, 1.0]:
        load_files = PATH + EXPERIMENT + '_label_*_dim_' + str(2) + '_std_' + str(RADIUS) + '_.pickle'
        read_list = glob.glob(load_files)
        list_of_dfs = []
        for file in read_list:
            with open(file, 'rb') as file:
                d = pickle.load(file)
                for column in d:
                    new = d[[column]].copy()
                    if column in PLOTCOL:
                        new['measure'] = PLOT_RENAME[column]
                    else:
                        new['measure'] = column
                    new.columns = new.columns.str.replace(column, 'Bottleneck distance')
                    list_of_dfs.append(new)

        big_df = pd.concat(list_of_dfs)
        big_df['Noise level'] = RADIUS
        list_of_bigdfs.append(big_df)
    bigbig_df = pd.concat(list_of_bigdfs)
    df_dict[EXPERIMENT + str(2)] = bigbig_df
    
for EXPERIMENT in ['fashion_mnist']:
    list_of_bigdfs = []
    for RADIUS in [0.001, 0.01, 0.1, 1.0]:
        load_files = PATH + EXPERIMENT + '_label_*_dim_' + str(3) + '_std_' + str(RADIUS) + '_.pickle'
        read_list = glob.glob(load_files)
        list_of_dfs = []
        for file in read_list:
            with open(file, 'rb') as file:
                d = pickle.load(file)
                for column in d:
                    new = d[[column]].copy()
                    if column in PLOTCOL:
                        new['measure'] = PLOT_RENAME[column]
                    else:
                        new['measure'] = column
                    new.columns = new.columns.str.replace(column, 'Bottleneck distance')
                    list_of_dfs.append(new)

        big_df = pd.concat(list_of_dfs)
        big_df['Noise level'] = RADIUS
        list_of_bigdfs.append(big_df)
    bigbig_df = pd.concat(list_of_bigdfs)
    df_dict[EXPERIMENT + str(3)] = bigbig_df
```
