```python
import numpy as np
import cv2
import pickle
import matplotlib.pyplot as plt
import time
import pandas as pd
import os
import glob
import seaborn as sns
```

```python
# Draw discriminator result
plot_df = pd.DataFrame()
cols = ['EMaP (d=2)', 'EMaP (d=3)', 'LIME +', 'LIME zero', 'SHAP']

DIM = 2
RADIUS = 0.00001
EXPERIMENT = 'mnist'
# EXPERIMENT = 'fashion_mnist'
PATH = 'result/discriminator/'
SAVEPATH = 'result/images/'
stat = ['True positive' , 'True negative']
mode = stat[1]
ytitle = mode
if mode == 'True positive':
    offset = 0
else:
    offset = 1  
    
# Load dim 2
load_files = PATH + EXPERIMENT + '_label_*_dim_' + str(DIM) + '_std_' + str(RADIUS) + '_.pickle'
read_list = glob.glob(load_files)
list_of_dfs = []
for file in read_list:
    with open(file, 'rb') as file:
        d = pickle.load(file)
    list_of_dfs.append(d)
big_df = pd.concat(list_of_dfs, ignore_index=True)

plot_df[cols[0]] = big_df[big_df.columns[4+offset]]

# Load dim 3
DIM = 3
load_files = PATH + EXPERIMENT + '_label_*_dim_' + str(DIM) + '_std_' + str(RADIUS) + '_.pickle'
read_list = glob.glob(load_files)
list_of_dfs = []
for file in read_list:
    with open(file, 'rb') as file:
        d = pickle.load(file)
    list_of_dfs.append(d)
big_df = pd.concat(list_of_dfs, ignore_index=True)

plot_df[cols[1]] = big_df[big_df.columns[4+offset]]

# Load others
DIM = 2
load_files = PATH + EXPERIMENT + '_label_*_dim_' + str(DIM) + '_std_' + str(RADIUS) + '_zero_base.pickle'
read_list = glob.glob(load_files)
list_of_dfs = []
for file in read_list:
    with open(file, 'rb') as file:
        d = pickle.load(file)
    list_of_dfs.append(d)
big_df = pd.concat(list_of_dfs, ignore_index=True)

plot_df[cols[2]] = big_df[big_df.columns[2+offset]] 
plot_df[cols[3]] = big_df[big_df.columns[0+offset]]
plot_df[cols[4]] = big_df[big_df.columns[4+offset]]
```

```python
sns.set_style("white")
# sns.set_context(rc = {'patch.linewidth': 10.0})
sns.set(font_scale=1.45)
height = 5
width = 8
plot_order = [cols[2],cols[3],cols[4],cols[0],cols[1]]
g = sns.catplot(x="variable", y="value", data=pd.melt(plot_df),
                kind="bar",   
#                 palette=sns.color_palette(['coral']),
#                 palette = 'rocket_r',
                palette = 'viridis_r',
#                 edgecolor = 'white',
                alpha=.99,
                height=height,
                linewidth=1,
                aspect=width/height,
                order= plot_order)
g.set_axis_labels("", "True positive")
g.set_ylabels(ytitle, fontsize = 20)
# plt.ylim(0.,1.01)
plt.tight_layout()
# plt.title("Discriminator performance on MNIST", fontsize = 24)
```
